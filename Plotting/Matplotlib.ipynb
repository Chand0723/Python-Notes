```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

There are two ways to create a plot using Matplotlib
1. Functional
2. Object oriented

```python
# Functional Method

x = np.linspace(0,10, 20)
y = x**2
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Square Plot')
plt.show()
```

```python
plt.subplot(1,2,1) #(row, column, plot number)
plt.plot(x,y,'r--')

plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
# Objected Oriented Method

fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #[left, bottom, width, height]
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Square Plot')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # percentage of axes1

axes1.plot(x,y)
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Large Plot')

axes2.plot(y,x,'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Small Plot')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
# axes[0].plot(x,y)
# axes[1].plot(y,x)
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
plt.tight_layout()
```

A common issue with matplolib is overlapping subplots or figures. We ca use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

```python
fig = plt.figure(figsize=(8,4), dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12,5))
axes[0].plot(x,y,'r')
axes[1].plot(y,x,'g')
plt.tight_layout()
```

Matplotlib can generate high-quality output in a number formats, including PNG, JPG, EPS, SVG, PGF and PDF. 

To save a figure to a file we can use the `savefig` method in the `Figure` class

```python
fig.savefig("Square.png", dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x^2")
ax.plot(x, x**3, label="x^3")
ax.legend()
```

The **legend** function takes an optional keyword argument **loc** that can be used to specify where in the figure the legend is to be drawn. The allowed values of **loc** are numerical codes for the various places the legend can be drawn.

http://matplotlib.org/users/legend_guide.html#legend-location

```python
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner

# .. many more options are available

# Most common to choose
ax.legend(loc=0) # let matplotlib decide the optimal location
```

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
